```python
# Python 3
# Get today's featured article from English Wikipedia

import datetime
import requests

today = datetime.datetime.now()
date = today.strftime('%Y/%m/%d')

url = 'https://api.wikimedia.org/feed/v1/wikipedia/en/featured/' + date

def read_token() -> str:
  with open("../data/access_token.txt", "r") as file:
    data = file.read()
  return data
token = read_token()

headers = {
  'Authorization': 'Bearer ' + token,
  'User-Agent': 'python_init'
}

response = requests.get(url, headers=headers)
resp = response.json()
print(resp)
```

```python
import pprint
pp = pprint.PrettyPrinter(indent=4)
pp.pprint(resp)
```

```python
most_read_articles = resp["mostread"]["articles"]

article_infos = []
for article in most_read_articles:
    article_dict = {
        "title" : article["title"],
        "description" : article["description"],
        "language": article["lang"],
        "views": article["views"],
        "url": article["content_urls"]["desktop"]["page"]
    }
    article_infos.append(article_dict)


```

```python
import pandas as pd

pd.DataFrame(article_infos)
```
