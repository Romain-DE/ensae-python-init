```python
import requests
import bs4

# Chansons de Paul McCartney
url = "https://genius.com/artists/Paul-mccartney/songs"
source = requests.get(url).text
soup = bs4.BeautifulSoup(source,'html.parser')
```

```python
# Beautiful Soup permet d'accder au code source d'une page web
soup
```

Il est commode d'inspecter le code source d'une page avec son navigateur d'abord afin de bien cibler ses requtes

```python
# ===== Point syntaxe CSS ======
# div reprsente une division de texte
# ul introduit une liste sans numrotation
# li introduit chaque item de la liste
# voir doc ici :  https://developer.mozilla.org/fr/docs/Web/HTML/Element/li


list_songs = soup.find('div',{'class':'ListSectiondesktop__Container-sc-53xokv-0'})\
.find('div', {'class': 'ListSectiondesktop__Content-sc-53xokv-7'})\
.find('ul', {'class': 'ListSectiondesktop__Items-sc-53xokv-8'})\
.find_all('li', {'class': 'ListItem__Container-sc-122yj9e-0'})
```

```python
def get_title(elt) -> str:
    return elt.find('div', {'class': 'ListItem__InfoContainer-sc-122yj9e-3'})\
        .find('h3', {'class': 'ListItem__Title-sc-122yj9e-4'})\
        .text

def get_stats(elt) -> str:
    return elt.find('div', {'class': 'ListItem__InfoContainer-sc-122yj9e-3'})\
        .find('div', {'class': 'ListItem__Info-sc-122yj9e-5'})\
        .find('span', {'class':'LabelWithIcon__Label-hjli77-1'})\
        .text

def get_url(elt) -> str:
    return elt.find('a', {'class': 'ListItem__Link-sc-122yj9e-1'}, href=True)["href"]

def get_credited_artists(elt) -> str:
    artist_and_song_name_from_url = get_url(elt).split("/")[-1].split("-")
    song_title = [word.lower() for word in get_title(elt).split(" ")]

    credited_artists = [word for word in artist_and_song_name_from_url if word not in song_title + ["lyrics"]]
    return " ".join(credited_artists).title()
    

for i in list_songs:
    title = get_title(i)
    stats = get_stats(i)
    artists = get_credited_artists(i)
    print(get_url(i))

```

```python
# Runion des informations
list_of_infos = []

for i in list_songs:
    title = get_title(i)
    stats = get_stats(i)
    artists = get_credited_artists(i)
    list_of_infos.append((title, stats, artists))
```

```python
# Synthse des informations dans un df
import pandas as pd

song_df = pd.DataFrame(list_of_infos, columns=["Title", "Views of lyrics", "Credited artists"])
song_df
```
